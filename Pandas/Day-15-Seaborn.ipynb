```python
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
sns.displot(df.displacement)
```

```python
sns.displot(df.horsepower)
```

```python
sns.displot(df.model_year)
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
sns.heatmap(data=df.corr())
```

```python
sns.heatmap(data=df.corr(),annot=True)
```
