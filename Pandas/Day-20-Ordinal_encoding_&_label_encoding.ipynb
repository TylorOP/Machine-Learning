```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/thispc/Desktop/Machine-learning/pandas/kaggle-dataset/customer.csv')
```

```python
df
```

```python
X = df[['review','education']] #ordinal data seperated
```

```python
X
```

```python
y = df['purchased']
```

```python
y
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
obj1 = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']],dtype=int)
```

```python
obj1
```

```python
ordinal_encode_X =obj1.fit_transform(X)
```

```python
X_encoded_df = pd.DataFrame(ordinal_encode_X,columns=['Review_encoded','Education'])
```

```python
pd.concat([X_encoded_df,df],axis=1)  #correct data encoded
```

```python
# label encodeing
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
obj2 = LabelEncoder()
```

```python
y_encnoded = obj2.fit_transform(y)
```

```python
y_encoded_df = pd.DataFrame(y_encnoded,columns=['y_encoded'])
```

```python
y_encoded_df
```

```python
# gender one-hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
obj = OneHotEncoder(sparse_output=False,dtype=int,drop='first') 
```

```python
array_gender = obj.fit_transform(df[['gender']])
```

```python
array_gender
```

```python
Encoded_gender = pd.DataFrame(array_gender,columns=['encoded_gender'])
```

```python
final_df = pd.concat([df,Encoded_gender,X_encoded_df,y_encoded_df,],axis=1)
```

```python
final_df
```

```python
final_df.columns
```

```python
final_df.drop(columns=['gender','review','education','purchased'],axis=1,inplace=True)
```
