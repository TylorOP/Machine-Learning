```python
X = df.drop(columns='Activity')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.30)
```

```python
X_train
```

```python
Var_obj = VarianceThreshold(threshold=0.20)
```

```python
Var_obj.fit(X_train)
```

```python
X_var_threshold_data_train = Var_obj.transform(X_train)
```

```python
X_var_threshold_data_test = Var_obj.transform(X_test)
```

```python
X.columns[Var_obj.get_support()]
```

```python
X_train_threshold_df = pd.DataFrame(X_var_threshold_data_train,columns=X.columns[Var_obj.get_support()])
```
