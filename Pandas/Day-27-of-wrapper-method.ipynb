```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/thispc/Desktop/Machine-learning/pandas/kaggle-dataset/Iris.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = df['Species']     
```

```python
le = LabelEncoder()
```

```python
y_encoded = le.fit_transform(y)
```

```python
X = df.drop(columns='Species')
```

```python
y = pd.DataFrame(pd.Series(y_encoded),columns=['y_encoded'])
```

```python
pip install mlxtend
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
```

```python
from sklearn.model_selection import cross_val_score  
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# As target data is in Categorical we use LogisticRegression model
# if target data is in continous we use linear regression
```

```python
lr =  LogisticRegression()
```

```python
efs = EFS(lr,scoring='accuracy',max_features=5,cv=3,n_jobs=-1)


#cv cross validation means divide the data in for eg 3 parts if there are 150 rows then 50,50,50 to avoid 
# biasness of data
```

```python
efs
```

```python
model = efs.fit(X,y.values.ravel())
```

```python
# Explanation:
# .values will give the values in a numpy array (shape: (n,1))
# .ravel will convert that array shape to (n, ) (i.e. flatten it)
```

```python
model.best_feature_names_
```

```python
model.best_score_
```

```python
model.subsets_       # exhaustive -> work on all possible combination
```

```python
X = df1.drop(columns=['survived'],axis=1)
X.drop(columns=['ticket'],inplace=True)
X
```

```python
y = df1['survived']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oe = OneHotEncoder(sparse_output=False,dtype=int)
```

```python
encoded =pd.DataFrame(oe.fit_transform(X[['embarked','sex']]),columns=['embarked_Q','Embarked_S','Embarked_C','sex_encode_M','Sex_encode_F'])

```

```python
#print(encoded.index)
print(X.index)
```

```python
final_X_encoded = pd.concat([encoded.reset_index(drop=True), X.reset_index(drop=True)], axis=1)
```

```python
#final_X_encoded.dropna(inplace=True)
final_X_encoded.drop(columns=['sex','embarked'],inplace=True)
```

```python
final_X_encoded
```

```python
encoded
y
```

```python
#exhaust
```

```python
from sklearn.model_selection import cross_val_score
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
efs = EFS(lr,scoring='accuracy',max_features=10,cv=3,n_jobs=-1)       #n_jobs=-1 means use all cpu cores
```

```python
efs
```

```python
model = efs.fit(final_X_encoded,y.values.ravel())
```

```python
model.best_score_
```

```python
model.feature_names
```

```python
model.subsets_
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
#both forward seletion and backward elimination
```

```python
le = LogisticRegression()
```

```python
sfs = SFS(le,
        k_features=7,
        forward=True,
        scoring='accuracy',
        cv=5,
        n_jobs=-1,)
```

```python
sfs
```

```python
model = sfs.fit(final_X_encoded,y)
```

```python
model.feature_names
```

```python
model.subsets_
```

```python
pd.DataFrame(model.get_metric_dict())
```
