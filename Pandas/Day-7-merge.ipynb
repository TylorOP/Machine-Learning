```python
emp = {
    'empid' : [105,102,108],
    'empname': ['shub','ram','rahul'],
    'city' : ['pune','solapur','bangalore']
}
```

```python
emp_hr = {
    'hr_id' : [201,105,108,873,463],
    'sal' : [4352,54635,4532,453,876],
    'dep' : ['IT','Admin','HR','Marketing','Sales'],
    'empname': ['go','goa','goan','goany','gunny']
    
}
```

```python
df = pd.DataFrame(emp)
df1 = pd.DataFrame(emp_hr)
```

```python
df
```

```python
df1
```

```python
df.merge(df1,on='empid') # error due to different column name # default inner join and same coumn name
```

```python
df.merge(df1,left_on='empid',right_on='hr_id') #different column name
```

```python
df.merge(df1,left_on='empid',right_on='hr_id',how='inner')
```

```python
df.merge(df1,left_on='empid',right_on='hr_id',how='left')
```

```python
df.merge(df1,left_on='empid',right_on='hr_id',how='right')
```

```python
df.merge(df1,left_on='empid',right_on='hr_id',how='outer')
```

```python
df.merge(df1,left_on='empid',right_on='hr_id',how='outer',suffixes=('_emp_data', '_hr_data'))
```

```python
df.merge(df1, how='cross')
```
