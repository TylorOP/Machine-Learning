# How would you group the data by the 'Category' column and calculate the total sales for each category?

```python
df['Category'].unique()
```

```python
obj = df.groupby('Category')
```

```python
#my answer #also correct
obj.agg({'Sales':[min,max,sum,]})
```

```python
#google
df.groupby(df['Category'])['Sales'].sum()  #ans
df.groupby(df['Category'])['Sales'].count()
df.groupby(df['Category'])['Sales'].mean()
df.groupby(df['Category'])['Sales'].median()
df.groupby(df['Category'])['Sales'].min()
df.groupby(df['Category'])['Sales'].max()
df.groupby(df['Category'])['Sales'].first()
df.groupby(df['Category'])['Sales'].last()
df.groupby(df['Category'])['Sales'].var()
df.groupby(df['Category'])['Sales'].describe()



```
