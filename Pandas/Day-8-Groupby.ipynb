# How would you normalize 'Sales' within each 'Region'?


##### Ans 
### to normalise we can use min-max method or z-score

### formula for 
# For Z-Score Normalization: (x - x.mean()) / x.std() computes the z-score.
# For Min-Max Normalization: (x - x.min()) / (x.max() - x.min()) scales the values between 0 and 1.

```python
df.groupby('Region')['Sales'].transform(lambda x : (x -x.mean()) / x.std())

#z-score
# this step process the data around mean and scale according to std,so that data falls in 
# bell curve

# ++++++++++++++++++++++++++++


df.groupby('Region')['Sales'].transform(lambda x : (x - x.min()) / (x.max() - x.min()))

#min-max

#  Min-Max normalization scales the data to a fixed range, usually between 0 and 1.
# where different features need to be compared on the same scale .
# Example: Number of Employees (1 to 10,000) vs. Rating (1 to 5).
# Different units can lead to one feature dominating the model. 
# Normalizing to 0-1 puts them on an equal footing.
```

# How do you apply a custom function to calculate the profit margin within each 'Category'?

```python
def custom_function(x):
    x['profit-margin % '] = (x['Profit']/x['Sales'])*100
    return x

df.groupby('Category').apply(custom_function)
```

# Analyze the sales performance across different 'Regions' by grouping by 'Region' and 'State.' Calculate the total and average 'Sales' and 'Profit' for each region. Identify any regions with a negative profit margin.

```python
df.groupby(['Region','State']).agg({'Sales':['sum','mean'],'Profit':['sum','mean']}
                                  ).reset_index()
```

# Above is Total Sales Total Profit Avg Sales and Avg Profit

```python
result = df.groupby(['Region','State']).agg({'Sales':['sum','mean'],'Profit':['sum','mean']}
                                  ).reset_index()


result.columns = ['Region','State','Total Sales','Avg Sales','Total Profit','Avg Profit']


result['Profit margin %'] = (result['Total Profit'] / result['Total Sales']) * 100

result[result['Profit margin %'] < 0]
```
