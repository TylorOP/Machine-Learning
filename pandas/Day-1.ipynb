```python
photo_col = df['image']
```

```python
import os
import requests

folder_name = 'shubham-photo'

if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
for i,url in enumerate(photo_col):
    response = requests.get(url)
    if response.status_code == 200:
        with open(os.path.join(folder_name, f'image_{i+1}.png'),'wb') as file:
            file.write(response.content)
         print(f'Successfully downloaded image_{i+1}.png')
              
    else:
        print(f"failed to download {url}")
        
```
